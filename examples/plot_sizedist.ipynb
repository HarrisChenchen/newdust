```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from newdust.graindist import sizedist
```

```python
MDTEST = 1.e-4 # g cm^-2
RHO1   = 3.0   # g cm^-3
RHO2   = 4.0
```

```python
gtest = sizedist.Grain(0.3)

print("For a = %.2f, rho = %.1f, nd = %.2e cm^-2" % (gtest.a, RHO1, gtest.ndens(MDTEST,RHO1)))
print("For a = %.2f, rho = %.1f, nd = %.2e cm^-2" % (gtest.a, RHO2, gtest.ndens(MDTEST,RHO2)))
```

```python
ptest = sizedist.Powerlaw()

ax = plt.subplot(111)
ptest.plot(ax, MDTEST, RHO1, label=r"$\rho$ = %.1f"%(RHO1))
ptest.plot(ax, MDTEST, RHO2, label=r"$\rho$ = %.1f"%(RHO2))
plt.legend(loc='upper left')
```

```python
etest = sizedist.ExpCutoff(nfold=5)

ax = plt.subplot(111)
etest.plot(ax, MDTEST, RHO1, label=r"$\rho$ = %.1f"%(RHO1))
etest.plot(ax, MDTEST, RHO2, label=r"$\rho$ = %.1f"%(RHO2))
plt.legend(loc='upper left')
```

## You can also get the mass density of dust grains

But the .plot function only does number density

```python
plt.plot(ptest.a, ptest.mdens(MDTEST, RHO1), label='Powerlaw')
plt.plot(etest.a, etest.mdens(MDTEST, RHO1), label='ExpCutoff')
plt.loglog()
plt.xlabel('Radius (um)')
plt.ylabel('Mass density (g cm$^{-2}$ um$^{-1}$)')
plt.legend(loc='upper right', frameon=False)
```
